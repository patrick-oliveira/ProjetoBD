```python
dados_globais = dados_globais.replace(np.nan, -1)
```

```python
paises_info = dados_globais[['iso_code', 'location', 'population', 'population_density', 'gdp_per_capita', 'aged_65_older', 'aged_70_older']].drop_duplicates()
paises_info = paises_info.sort_values('iso_code').set_index(pd.Series(range(100, 100 + 215)))

paises_index = list(paises_info.index)
paises_code  = list(paises_info['iso_code'])
code_to_index = {code:index for code, index in zip(paises_code, paises_index)}

informacoes_globais_mortes = dados_globais[['iso_code', 'date', 'new_deaths', 'total_deaths', 'new_deaths_per_million', 'total_deaths_per_million']]
informacoes_globais_casos = dados_globais[['iso_code', 'date', 'new_cases', 'total_cases', 'new_cases_per_million', 'total_cases_per_million']]
informacoes_globais_testes = dados_globais[['iso_code', 'date', 'new_tests', 'new_tests_per_thousand', 'total_tests', 'total_tests_per_thousand', 'positive_rate', 'tests_per_case']]
informacoes_globais_vacinacoes = dados_globais[['iso_code', 'date', 'new_vaccinations', 'total_vaccinations', 'total_vaccinations_per_hundred']]
informacoes_globais_hospitalizacoes = dados_globais[['iso_code', 'date', 'hosp_patients', 'icu_patients', 'icu_patients_per_million', 'weekly_icu_admissions', 'weekly_hosp_admissions']]

informacoes_globais_mortes.loc[:, 'new_deaths'] = informacoes_globais_mortes.loc[:, 'new_deaths'].astype(int)
informacoes_globais_mortes.loc[:, 'total_deaths'] = informacoes_globais_mortes.loc[:, 'total_deaths'].astype(int)

informacoes_globais_casos.loc[:, 'new_cases'] = informacoes_globais_casos.loc[:, 'new_cases'].astype(int)
informacoes_globais_casos.loc[:, 'total_cases'] = informacoes_globais_casos.loc[:, 'total_cases'].astype(int)

informacoes_globais_testes.loc[:, 'new_tests'] = informacoes_globais_testes['new_tests'].astype(int)
informacoes_globais_testes.loc[:, 'total_tests'] = informacoes_globais_testes['total_tests'].astype(int)

informacoes_globais_vacinacoes.loc[:, 'new_vaccinations'] = informacoes_globais_vacinacoes['new_vaccinations'].astype(int)
informacoes_globais_vacinacoes.loc[:, 'total_vaccinations'] = informacoes_globais_vacinacoes['total_vaccinations'].astype(int)

informacoes_globais_hospitalizacoes.loc[:, 'hosp_patients'] = informacoes_globais_hospitalizacoes['hosp_patients'].astype(int)
informacoes_globais_hospitalizacoes.loc[:, 'icu_patients'] = informacoes_globais_hospitalizacoes['icu_patients'].astype(int)

tables = [informacoes_globais_mortes, informacoes_globais_casos, informacoes_globais_testes, \
          informacoes_globais_vacinacoes, informacoes_globais_hospitalizacoes]
table_names = ['informacoes_globais_mortes', 'informacoes_globais_casos', 'informacoes_globais_testes', \
               'informacoes_globais_vacinacoes', 'informacoes_globais_hospitalizacoes']



for df in tables:
    df['iso_code'] = df.loc[:, 'iso_code'].apply(lambda x: code_to_index[x])
    cols = list(df.columns)
    cols[0] = 'code'
    df.columns = cols

for df, name in zip(tables, table_names):
    df = df.replace(-1, np.nan)
    df.to_csv('data/tabelas/{}.csv'.format(name), index = False, header = False)
```

```python
# baixando os dados de http://cnes.datasus.gov.br/pages/downloads/arquivosBaseDados.jsp

import shutil
import urllib.request as request
from contextlib import closing

base_de_dados = ['BASE_DE_DADOS_CNES_202101.ZIP', 'BASE_DE_DADOS_CNES_202102.ZIP', 'BASE_DE_DADOS_CNES_202103.ZIP']

for dados in base_de_dados:
    with closing(request.urlopen('ftp://ftp.datasus.gov.br/cnes/{}'.format(dados))) as r:
        with open(dados, 'wb') as f:
            shutil.copyfileobj(r, f)
```

```python
populacao_municipios = pd.read_csv('data/populacao_municipios.csv', encoding = 'latin1')

cod_munic = list(populacao_municipios['COD. MUNIC'].astype(str))
cod_uf    = list(populacao_municipios['COD. UF'].astype(str))

cod_munic_completo = []

for id_uf, id_munic in zip(cod_uf, cod_munic):
    complemento = ''
    while (len(id_uf) + len(id_munic) + len(complemento)) < 7:
        complemento = '0'+complemento
    
    cod_munic_completo.append(id_uf + complemento + id_munic)
    
populacao_municipios['COD. MUNIC'] = cod_munic_completo

populacao_municipios = populacao_municipios[['COD. MUNIC', ' POPULAÇÃO ESTIMADA ']]
populacao_municipios.columns = ['id', 'populacao']
```

```python
import re
def f(x):
    x = x.replace(',', '')
    x = x.replace('.', '')
    x = re.sub('\((.*?)\)', '', x)
    return x
```

```python
populacao_municipios['populacao'] = populacao_municipios['populacao'].apply(f)
populacao_municipios['populacao'] = populacao_municipios['populacao'].astype(int)
populacao_municipios.head()
```
